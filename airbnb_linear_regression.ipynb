## Linear Regression

```python
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
def LRfindAveragePrice(location):
    return (df[(df.neighbourhood_cleansed == location)])

area = LRfindAveragePrice('Allerton')
area.drop(['neighbourhood_cleansed'], axis=1)
area.info()
```

```python
reg = linear_model.LinearRegression()
reg.fit(area[['accommodates', 'beds', 'bedrooms', 'bathrooms', 'bed_type', 'property_type']], area.price)
#fit == training a model with the available points
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[4, 4, 1, 1, 1, 1]])
```

```python
X_train, X_test, y_train, y_test = train_test_split(area[['accommodates', 'beds', 'bedrooms', 'bedrooms', 'bed_type', 'property_type']],
    area['price'], test_size = 0.3, random_state=42)
reg_all = LinearRegression()
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)

reg_all.score(X_test, y_test)
```
