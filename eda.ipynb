```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("s3://datasetairbnb/airbnb_cleaned_winter.csv")
```

```python
df
```

```python
df.drop('bed_type', axis=1, inplace=True)
```

## EDA

```python
def findAveragePrice(location):
    return (df[(df.neighbourhood_cleansed == location)])

area = findAveragePrice('Allerton')
area.plot.scatter('price', 'accommodates')
```

```python
area_describe = area.describe().T
area_describe
```

```python
area_describe['iqr_max'] = area_describe['75%'] + 1.5 * (area_describe['75%'] - area_describe['25%'])
area_describe['iqr_min'] = area_describe['75%'] - 1.5 * (area_describe['75%'] - area_describe['25%'])

cols_to_drop_outliers = ['minimum_nights', 'price','accommodates', 'beds', 'bedrooms', 'bathrooms','number_of_reviews', 'review_scores_rating']

for col in cols_to_drop_outliers:
    print(col)
    df = df[((df[col] > area_describe.loc[col, 'iqr_min']) & (df[col] < area_describe.loc[col, 'iqr_max']))]
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(area.corr(), annot=True, fmt=".2f");
plt.show()
```
