```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import codecs, json 

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('s3://datasetairbnb/airbnb_cleaned_summer.csv')
```

```python
df.info()
```

```python
list = []
```

```python
for i in df.neighbourhood_cleansed:
    if i not in list:
        list.append(i)
```

```python
list.sort()
```

```python
list
```

```python
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
list
```

```python
def findAveragePrice(location):
    return (df[(df.neighbourhood_cleansed == location)])

res = {}

# res[i]
def getall():
    for i in list:
        res[i] = findAveragePrice(i)

    coefficients = {}

    for i in res:
        regStandardize = linear_model.LinearRegression()
        regStandardize.fit(res[i][['accommodates', 'beds', 'bedrooms', 'bathrooms', 'bed_type', 'property_type',
                                   'number_of_reviews', 'review_scores_rating']], res[i].price)
        
        keys = ['accommodates', 'beds', 'bedrooms', 'bathrooms', 'bed_type', 'property_type', 
                'number_of_reviews', 'review_scores_rating']
        coefficients[i] = dict(zip(keys, regStandardize.coef_.tolist()))
        coefficients[i]['intercepts'] = regStandardize.intercept_

    for i in coefficients:
        coefficients[i]

    return coefficients
test = getall()
```
